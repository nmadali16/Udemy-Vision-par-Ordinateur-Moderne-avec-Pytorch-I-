```
import cv2
import numpy as np
import matplotlib.pyplot as plt
import torch
```

```


# Charger l'image avec OpenCV
image_path = 'monalisa.jpg'  # Remplacez par le chemin de votre image
image = cv2.imread(image_path)
```

```
# Afficher l'image avec Matplotlib
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Image originale")
plt.show()
```

```
# Afficher la taille de l'image
print(f"Taille de l'image originale : {image.shape}")

# Redimensionner l'image
new_size = (224,224)  # Remplacez par la nouvelle taille souhaite
resized_image = cv2.resize(image, new_size)


```

```
# Afficher la nouvelle taille de l'image redimensionne
print(f"Taille de l'image redimensionne : {resized_image.shape}")
# Afficher l'image redimensionne avec Matplotlib
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
plt.title("Image redimensionne")
plt.show()
```

```
# Convertir l'image en niveaux de gris avec OpenCV
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Afficher l'image en niveaux de gris avec Matplotlib
plt.imshow(gray_image, cmap='gray')
plt.title("Image en niveaux de gris")
plt.show()

# Afficher la taille de l'image en niveaux de gris
print(f"Taille de l'image en niveaux de gris : {gray_image.shape}")
```
