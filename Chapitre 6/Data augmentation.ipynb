```python
import torch
import matplotlib.pyplot as plt
import torchvision.transforms as transforms
from PIL import Image
```

```python
# Load an image using PIL
input_image_path = 'monalisa.jpeg'  # Replace 'path_to_your_image.jpg' with the actual path
original_img = Image.open(input_image_path)
```

```python
transform = transforms.Compose([
    transforms.ToTensor(),# 255 0,1


])
```

```python
transform(original_img).shape # 0,1,2  1,2,0
```

```python
transform(original_img).permute(1, 2, 0).shape
```

```python

plt.title('Original Image')
plt.imshow(transform(original_img).permute(1, 2, 0), cmap='gray')
plt.axis('off')
plt.show()
```

```python

# Application de différentes augmentations de données à l'aide de torchvision
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.RandomHorizontalFlip(),

])

# Appliquer les augmentations et afficher les images transformées
transformed_img = transform(original_img).permute(1, 2, 0)


plt.imshow(transformed_img, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python

# Application de différentes augmentations de données à l'aide de torchvision
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.RandomVerticalFlip(),

])

# Appliquer les augmentations et afficher les images transformées
transformed_img = transform(original_img).permute(1, 2, 0)


plt.imshow(transformed_img, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python

# Application de différentes augmentations de données à l'aide de torchvision
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.RandomAffine(degrees=0, translate=(0.1, 0.1))

])

# Appliquer les augmentations et afficher les images transformées
transformed_img = transform(original_img).permute(1, 2, 0)


plt.imshow(transformed_img, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python

# Application de différentes augmentations de données à l'aide de torchvision
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.RandomAffine(degrees=-30, translate=(0, 0))

])

# Appliquer les augmentations et afficher les images transformées
transformed_img = transform(original_img).permute(1, 2, 0)


plt.imshow(transformed_img, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
